```python
import pandas as pd
import requests
import json
```

```python
# First get count of total pages
```

```python
url_template = 'https://www.woolsboutiqueuomo.com/collections/all/products.json?page={}'

page = 1
while True:
    
    # Create a session and set headers
    session = requests.Session()
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.182 Safari/537.36',
        'Accept-Language': 'en-US,en;q=0.9',
        # Add other necessary headers here
    }
    session.headers.update(headers)
    
    url = url_template.format(page)
    
    response = requests.get(url)
    product_data = response.json()

    # If the product data is empty, it means we have reached the last page
    if not product_data["products"]:
        break

    page += 1

total_pages = page - 1
print("Total Pages:", total_pages)
```

```python
# next title for each product
```

```python
title = []

url_template = 'https://www.woolsboutiqueuomo.com/collections/all/products.json?page={}'

for j in range(total_pages):
    
    # Create a session and set headers
    session = requests.Session()
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.182 Safari/537.36',
        'Accept-Language': 'en-US,en;q=0.9',
        # Add other necessary headers here
    }
    
    session.headers.update(headers)
    
    url = url_template.format(j)
    # Send a GET request to the URL
    response = requests.get(url)

    # Extract the JSON data from the response
    product_data = response.json()
    
    for i in range(len(product_data["products"])):
        product = product_data["products"][i]
        title.append(product["title"])
```

```python
# get the link address for each product using 'handle'
```

```python
link = []

url_template = 'https://www.woolsboutiqueuomo.com/collections/all/products.json?page={}'

for j in range(total_pages):
    
    # Create a session and set headers
    session = requests.Session()
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.182 Safari/537.36',
        'Accept-Language': 'en-US,en;q=0.9',
        # Add other necessary headers here
    }
    
    session.headers.update(headers)
    
    url = url_template.format(j)
    # Send a GET request to the URL
    response = requests.get(url)

    # Extract the JSON data from the response
    product_data = response.json()
    
    for i in range(len(product_data["products"])):
        product = product_data["products"][i]
        link.append('https://www.woolsboutiqueuomo.com/products/{}'.format(product["handle"]))
```

```python
# do text preprocessing using nltk library on the list of titles
# this includes getting rid of stop-words, lower casing the the words and lemmatizing it
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from sklearn.metrics.pairwise import cosine_similarity

title2 = []

for text in title:
    
    stop_words = set(stopwords.words('english'))
    lemmatizer = WordNetLemmatizer()
    tokens = word_tokenize(text.lower())
    tokens = ' '.join([lemmatizer.lemmatize(token) for token in tokens if token.isalnum()])
    title2.append(tokens)

title2[:20]
```

```python
# convert all the list elements i.e title into a matrix of words where 1 indicates the word each element contains 
# and 0 means it does not contains
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vectorizer = CountVectorizer()
vector_matrix = count_vectorizer.fit_transform(title2)
vector_matrix
```

```python
tokens = count_vectorizer.vocabulary_.keys()
tokens
```

```python
vector_matrix.toarray()
```

```python
import pandas as pd

def create_dataframe(matrix, tokens):

    doc_names = [f'title_{i+1}' for i, _ in enumerate(matrix)]
    df = pd.DataFrame(data=matrix, index=doc_names, columns=tokens)
    return(df)

create_dataframe(vector_matrix.toarray(),tokens)
```

```python
# now compute the cosine similarity which tells us how similar each title of a product is to the tiltle of another product.
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_similarity_matrix = pd.DataFrame(cosine_similarity(vector_matrix))
cosine_similarity_matrix
```

```python
def group_similar_elements(similarity_matrix):
    N = len(similarity_matrix)
    groups = []

    visited = set()

    for i in range(N):
        if i not in visited:
            group = [i]
            visited.add(i)

            for j in range(i+1, N):
                if similarity_matrix[i][j] > 0.5:   # threshold of 0.5 is selected for similarity. 
                    group.append(j)                 # for more stricture on can increase the threshold parameter.
                    visited.add(j)

            if len(group) > 1:
                groups.append(group)

    return groups


similar_index = group_similar_elements(cosine_similarity_matrix)
similar_index[:10]
```

```python
similar_titles = []
for j in range(len(similar_index)):
    s = []
    for i in similar_index[j]:
        s.append(title[i])
    similar_titles.append(s)
    
similar_titles[:10]
```

```python
similar_titles_link = []
for j in range(len(similar_index)):
    s = []
    for i in similar_index[j]:
        s.append(link[i])
    similar_titles_link.append(s)
    
similar_titles_link[:10]
```

```python
json_data = []
for group,l in zip(similar_titles_link,range(len(similar_titles_link))):
    json_group = {f"product alternates{l}": []}
    for element_index in group:
        json_group[f"product alternates{l}"].append(element_index)
    json_data.append(json_group)
```
